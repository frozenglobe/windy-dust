```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm, rcParams, ticker
from scipy.optimize import curve_fit
```

Mia had a question about scipy curve_fit

```python
rcParams["figure.figsize"] = (4, 2)
model = lambda x, a, b: a * np.exp(-b * x)

x_data = np.arange(0, 10, 0.1)
y_data = model(x_data, 2.5, 1.3) + 0.2*np.random.normal(size=x_data.size)

cf, er = curve_fit(model, x_data, y_data)

plt.scatter(x_data, y_data, label='data', color='C0', s=5, marker='x')
plt.plot(x_data, model(x_data, *cf), label='a='+str(round(cf[0],1))+', b='+str(round(cf[1],1)), color='C3')
plt.legend(loc='upper right', prop={'size': 10})
```
