```python
import numpy as np
from astropy.io import fits
import stpsf
import matplotlib.pyplot as plt
import os
```

```python
os.mkdir('psf_fits')
```

### NIRCam

```python
nrc = stpsf.NIRCam()

def psf_hdu(filter:str):
    nrc.filter = filter
    psf = nrc.calc_psf(oversample=4)
    psf.writeto(f'psf_fits/{filter}_psf.fits', overwrite=True)
    return psf
```

```python
psf_hdu('F200W')
psf_hdu('F444W')
```

### MIRI

```python
miri = stpsf.MIRI()

def psf_hdu(filter:str):
    miri.filter = filter
    psf = miri.calc_psf(oversample=4)
    psf.writeto(f'psf_fits/{filter}_psf.fits', overwrite=True)
    return psf
```

```python
psf_hdu('F770W')
psf_hdu('F1280W')
psf_hdu('F2100W')
```

```python
stpsf.display_psf(fits.open('F1280W_psf.fits'), ext=3)
```
